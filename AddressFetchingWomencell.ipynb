```
import numpy as np
import random
import pandas as pd
import random as rd
from google.colab import files
from shapely.geometry import Polygon, Point


poly = Polygon([(28.881426, 77.075359), (28.856170, 77.211315), (28.417491, 77.166959),(28.582837, 76.848356),(28.822546, 76.952726)])

def random_points_within(poly, num_points):
    min_x, min_y, max_x, max_y = poly.bounds

    points = []

    while len(points) < num_points:
        random_point = Point([random.uniform(min_x, max_x), random.uniform(min_y, max_y)])
        if (random_point.within(poly)):
            points.append(random_point)

    return points


points = random_points_within(poly,1175)
lx=[]
ly=[]
crimelisttype=["","molestation","dowry","rape"]
cr=[]
for p in points:
    lx.append(p.x)
    ly.append(p.y)
    cr.append(rd.choice(crimelisttype))

dictionarydb={'latitude':lx,'longitude':ly,'crimeActivity':cr}
df = pd.DataFrame(dictionarydb)
df.to_csv('file1.csv') 
df.columns
df.groupby('crimeActivity').size() # finding number of occurences

#files.download('file1.csv')
```

```
rslt_df0 = df.loc[df['crimeActivity']=='dowry'] 
rslt_df1 = df.loc[df['crimeActivity']=='eve teasing'] 
rslt_df2 = df.loc[df['crimeActivity']=='molestation'] 
rslt_df3 = df.loc[df['crimeActivity']=='rape'] 

```

```
df2=pd.DataFrame(rslt_df1)#redefining of data set
df2.reset_index(drop=True)
rslt_df0.reset_index(drop=True)
rslt_df2.reset_index(drop=True)
rslt_df3=rslt_df3.reset_index(drop=True)
train3c=rslt_df3.head((rslt_df3.shape[0]*75)//100)
train2c=rslt_df2.head((rslt_df2.shape[0]*75)//100)
train1c=rslt_df1.head((rslt_df1.shape[0]*75)//100)
train0c=rslt_df0.head((rslt_df3.shape[0]*75)//100)

```

```
Traindata = pd.concat([train0c,train1c,train2c,train3c], axis=0)

Traindata=Traindata.reset_index(drop=True)
Ylabel=Traindata.iloc[:,-1]
Traindata=Traindata.drop(columns=['crimeActivity'])
Traindata
```

```
TestData=rslt_df0.tail((rslt_df3.shape[0]*25)//100)
TestData=TestData.drop(columns=['crimeActivity'])

```

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(Traindata, Ylabel)
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
                     metric_params=None, n_jobs=None, n_neighbors=5, p=2,
                     weights='uniform')
print("predictions:")
print(list(knn.predict(TestData)).count("rape"))
```
